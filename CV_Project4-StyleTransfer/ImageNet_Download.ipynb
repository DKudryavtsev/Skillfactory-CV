```python
# Training files
tf_record_pattern = os.path.join(
    'gs://mobile-ml-wg/tfrecord/imagenet2012/', '%s-*' % 'train')
data_files = tf.io.gfile.glob(tf_record_pattern)

# Validaion files
#tf_record_pattern = os.path.join(
#    'gs://mobile-ml-wg/tfrecord/imagenet2012/', '%s-*' % 'validation')
#data_files = tf.io.gfile.glob(tf_record_pattern)

n = 5  # Number of ImageNet TFRecord files to download
for i in range(n):
    file = tf.data.TFRecordDataset(data_files[i])
    fname = '/content/drive/MyDrive/ImageNet/' + data_files[i].split('/')[-1]
    writer = tf.data.experimental.TFRecordWriter(fname)
    writer.write(file)
```
